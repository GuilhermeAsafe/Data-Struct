```python
def entrada():
    a = int(input("Digite o valor de a: "))
    b = int(input("Digite o valor de b: "))
    c = int(input("Digite o valor de c: "))
    return a, b, c
```

```python
#i) transitividade se a|b e b|c -> a|c, Ã© verdadede
#sempre verdadeiro 
def DIVA(b,a):
    return (a % b == 0)
```

```python
#sempre verdadeiro
def DIVB(c,b):
    return (b%c == 0)
```

```python
def DIVC(c,a):
    return (a%c == 0)
```

```python
def comprova_transitividade():
    a, b, c = entrada()
    return (not((DIVA(b,a) and DIVB(c,b)) or DIVC(c,b))) #a -> b ~ not(a) or b
```
