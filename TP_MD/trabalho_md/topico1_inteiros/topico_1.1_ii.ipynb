```python
def entrada():
    a = int(input("Digite o valor de a: "))
    b = int(input("Digite o valor de b: "))
    c = int(input("Digite o valor de c: "))
    return a, b, c
```

```python
#i) transitividade se b|a e b|c -> b|c + a, Ã© verdadede
#sempre verdadeiro 
def DIVA(b,a):
    if (b % a == 0):
        return True
    else:
        return False
    
```

```python
#sempre verdadeiro
def DIVB(c,b):
    if (b%c == 0):
        return True
    else:
        return False
```

```python
def DIV_sum(a,b,c):
    if (b%(a+c) == 0):
        return True
    else:
        return False
        
```

```python
def comprova_ii():
    a, b, c = entrada()
    if (not(DIVA(b,a) and DIVB(c,b)) or DIV_sum(a,b,c)):
     return True
    else:
        return False
```

```python
print(comprova_ii())
```
