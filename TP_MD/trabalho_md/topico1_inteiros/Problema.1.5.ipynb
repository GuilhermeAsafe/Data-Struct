```python
def soma_divisores_proprios(n):
    soma = 0
    for d in range(1, n):
        if n % d == 0:
            soma += d
    return soma

limite = 1000
numeros_perfeitos = []

for n in range(1, limite):
    if soma_divisores_proprios(n) == n:
        numeros_perfeitos.append(n)

print(f"Números perfeitos menores que {limite}: {numeros_perfeitos}")
```
