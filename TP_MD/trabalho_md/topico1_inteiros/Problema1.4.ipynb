```python
n = 48
D48 = [d for d in range(1, n + 1) if n % d == 0]
tau_48 = len(D48)
sigma_48 = sum(D48)

print(f"Para n = {n}:")
print(f"Número de divisores (tau): {tau_48}")
print(f"Soma dos divisores (sigma): {sigma_48}\n")

m = 150
D150 = [d for d in range(1, m + 1) if m % d == 0]
tau_150 = len(D150)
sigma_150 = sum(D150)

print(f"Para m = {m}:")
print(f"Número de divisores (tau): {tau_150}")
print(f"Soma dos divisores (sigma): {sigma_150}")
```
