```python
def gcd_euclid_detalhado(a, b):
    a_original = a
    b_original = b
    
    if b > a:
        a, b = b, a
    
    print(f"Calculando MDC({a_original}, {b_original}) usando o Algoritmo de Euclides:")
    
    while b != 0:
        q = a // b
        r = a % b
        print(f"{a} = {q} * {b} + {r}")
        a = b
        b = r
    
    print(f"MDC final: {a}")
    return a

# 1. Verificar o exemplo: gcd(1057, 315)
print("\n--- Verificação do Exemplo (mdc(1057, 315)) ---")
gcd_euclid_detalhado(1057, 315)

# 2. Executar o programa modificado para 4181 e 2584
print("\n--- Cálculo do MDC(4181, 2584) ---")
gcd_euclid_detalhado(4181, 2584)
```

```python
#Resposta 1.7
#Por que é o pior caso: Porque os números de Fibonacci consecutivos forçam o resto (r) a ser o maior possível em relação ao próximo divisor (b). 
Isso faz com que o algoritmo demore o máximo de passos para chegar ao MDC. O teorema de Lamé prova que este é o caso que exige mais iterações.
```
