```python
n = 48

 
D48 = [d for d in range(1, n + 1) if n % d == 0]
print(f"Número: {n}")
print(f"Lista de divisores (D{n}): {D48}")

 
tau_48 = len(D48)
print(f"Número de divisores (tau({n})): {tau_48}")

 
sigma_48 = sum(D48)
print(f"Soma dos divisores (sigma({n})): {sigma_48}\n")

 
m = 150

# 1. Calcular a lista de divisores (D150)
Dm = [d for d in range(1, m + 1) if m % d == 0]
print(f"Número: {m}")
print(f"Lista de divisores (D{m}): {Dm}")

# 2. Calcular o número de divisores (tau(150))
tau_150 = len(Dm)
print(f"Número de divisores (tau({m})): {tau_150}")

# 3. Calcular a soma de todos os divisores (sigma(150))
sigma_150 = sum(Dm)
print(f"Soma dos divisores (sigma({m})): {sigma_150}")
```
