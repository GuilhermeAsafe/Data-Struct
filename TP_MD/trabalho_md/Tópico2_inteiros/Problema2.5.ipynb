```python
import math

def estimar_primos_no_intervalo(a, p):
    if p <= 1:
        pi_p = 0
    else:
        pi_p = p / math.log(p)
        
    a_ajustado = max(1, a - 1)
    if a_ajustado <= 1:
        pi_a_menos_1 = 0
    else:
        pi_a_menos_1 = a_ajustado / math.log(a_ajustado)
    
    estimativa_primos = pi_p - pi_a_menos_1
    
    print(f"Estimativa de primos no intervalo [{a}, {p}]:")
    print(f"pi({p}) ≈ {pi_p:.0f}")
    print(f"pi({a-1}) ≈ {pi_a_menos_1:.0f}")
    print(f"Número de primos (aprox.): {estimativa_primos:.0f}")
    
    return estimativa_primos

# Exemplo: Estimando primos no intervalo [100, 200]
estimar_primos_no_intervalo(100, 200)
```
