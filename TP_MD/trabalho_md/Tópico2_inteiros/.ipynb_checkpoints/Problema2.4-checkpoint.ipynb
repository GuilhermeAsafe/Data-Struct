```python
import math

def estimar_primos_na_faixa(tamanho_bits):
    t = tamanho_bits
    fator_constante = 1 / (2 * t * math.log(2))
    
    # Primos na faixa ≈ (fator_constante) * 2^t
    print(f"Para {t} bits:")
    print(f"Estimativa: ~ {fator_constante:.4f} * 2^{t}")
    
    # Para notação científica (visualização)
    log10_resultado = math.log10(fator_constante) + t * math.log10(2)
    expoente = math.floor(log10_resultado)
    mantissa = 10**(log10_resultado - expoente)
    
    print(f"Estimativa em notação científica: ~ {mantissa:.2f}e{expoente}\n")
    
tamanhos = [1024, 1536, 2048]

for t in tamanhos:
    estimar_primos_na_faixa(t)
```
