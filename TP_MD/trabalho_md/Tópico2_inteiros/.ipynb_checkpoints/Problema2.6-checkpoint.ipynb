```python
import random

 
def fermat_primality_test(n, k=10):
    
    if n < 2:
        return False
    elif n < 4:
        return True  # 2 e 3 são primos
    elif n % 2 == 0:
        return False # Números pares >2 são compostos

    for _ in range(k):
       
        a = random.randint(2, n - 2)
        
        if pow(a, n - 1, n) != 1:
            return False  
            
    return True  

 
NOME_CODIFICADO_84 = "072109120805181305270119010605270405271909172105091801271305091805120519"

 
DIGITOS_ALEATORIOS_16 = "3014986257008713" 

# Construção do número n de 100 dígitos como string
n_str = NOME_CODIFICADO_84 + DIGITOS_ALEATORIOS_16

 
n = int(n_str)

  
k = 10 
resultado = fermat_primality_test(n, k)

 
print("-" * 50)
print(f"Nome Original: GUILHERME ASAFE DE SIQUEIRA MEIRELES")
print(f"Número N (99/100 dígitos): {n_str}")
print(f"Número de dígitos de N: {len(str(n))}")
print("-" * 50)
print(f"Resultado do Teste de Primalidade de Fermat (k={k}):")

if resultado:
    print(">>> True (PROVAVELMENTE PRIMO) <<<")
    print("Fim do exercício.")
else:
    print(">>> False (DEFINITIVAMENTE COMPOSTO ou Pseudoprimo) <<<")
    print("Ação Necessária: Altere a variável DIGITOS_ALEATORIOS_16 e rode novamente.")

print("-" * 50)
```
