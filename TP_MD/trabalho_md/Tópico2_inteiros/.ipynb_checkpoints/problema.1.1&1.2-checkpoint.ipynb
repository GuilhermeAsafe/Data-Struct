```python
def eh_primo(n):
    if n <= 1:
        return False
    
    # Testar divisibilidade de 2 até n-1 (ou até int(n**0.5) + 1 para otimizar)
    for i in range(2, n):
        if n % i == 0:
            return False
    
    return True

# Teste com alguns números
n1 = 17
n2 = 18
n3 = 2
n4 = 1

print(f"O número {n1} é primo? {eh_primo(n1)}")
print(f"O número {n2} é primo? {eh_primo(n2)}")
print(f"O número {n3} é primo? {eh_primo(n3)}")
print(f"O número {n4} é primo? {eh_primo(n4)}")
```
