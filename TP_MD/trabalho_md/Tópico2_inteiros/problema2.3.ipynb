```python
n1 = 257
n2 = 65537

# O comando bin() converte o inteiro para uma string binária (prefixada com '0b')
binario_n1 = bin(n1)
binario_n2 = bin(n2)

print(f"Número: {n1}")
print(f"Representação binária: {binario_n1}")
print(f"Representação binária (sem '0b'): {binario_n1[2:]}\n")

print(f"Número: {n2}")
print(f"Representação binária: {binario_n2}")
print(f"Representação binária (sem '0b'): {binario_n2[2:]}")
```
